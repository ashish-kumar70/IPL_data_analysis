# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
```

# Import data

```python
file_path='G:/project1/completed/IPL data analysis/csv_files'
dataframes=[]
num=0
for root,dir,files in os.walk(file_path):
    for file in files:
        if file.endswith('.csv'):
            file_p=os.path.join(root,file)
            df=pd.read_csv(file_p)
            dataframes.append(df)
            print(f'{num} - {file}')
            num+=1
```

# cities in which most matches played

```python
Count_Match_location=Ipl_matches.groupby('City').agg({'City':'count'})
Count_Match_location.rename(columns={'City':'Matches_Played'},inplace=True)
Count_Match_location.sort_values(by='Matches_Played',ascending=False,inplace=True)
top_city_by_match=Count_Match_location.head(10)
top_city_by_match
```

```python
sns.lineplot(x='City',y='Matches_Played',data=top_city_by_match,marker='o')
plt.title("Total Matches Played In city")
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

# Most wickets taken by a player

```python
ball_by_ball_match.kind.unique()
```

```python
dismissial_balls=ball_by_ball_match[ball_by_ball_match.kind.isin([
    'caught','caught and bowled','bowled','stumped','lbw','hit wicket'])]
```

```python
dismissial_balls
```

```python
total_wickets=dismissial_balls.groupby('bowler').agg({'ID':'nunique','isWicketDelivery':'sum'})
total_wickets.rename(columns={'ID':'Matches_played','isWicketDelivery':'Total_wickets'},inplace=True)
total_wickets.sort_values(by='Total_wickets',ascending=False,inplace=True)
total_wickets.head(10)
```

```python
most_wickets=total_wickets.head(10)
```

# Most Sixes by a player

```python
six_hit_balls=ball_by_ball_match[ball_by_ball_match.batsman_run.isin([6])]
total_six_by_batsman = six_hit_balls.groupby('batter').size().reset_index(name='Total sixes')
top_ten_six_hitters=total_six_by_batsman.sort_values(by='Total sixes',ascending=False).head(10)
```

```python
ax = plt.axes()
ax.set_facecolor('silver')
plt.bar(x=top_ten_six_hitters["batter"],height=top_ten_six_hitters['Total sixes'],
        color=(0.1, 0.1, 0.1, 0.1),  edgecolor='cyan')
plt.grid(True)
plt.title("Most Sixes In IPL")
plt.xticks(rotation=60)
plt.show()
```

# Trophies Won

```python
final_matches=Ipl_matches[Ipl_matches.MatchNumber.isin(['Final'])]
final_matches
```

```python
trophies_won=final_matches.groupby('WinningTeam').agg({'ID':'count'}).sort_values(by='ID',ascending=False)
trophies_won.reset_index(inplace=True)
trophies_won.rename(columns={'WinningTeam':"Team",'ID':"Total Trophies won"},inplace=True)
trophies_won
```

```python
myexplode=[0.1,0,0,0,0,0,0]
plt.pie(trophies_won['Total Trophies won'],labels=trophies_won['Team'],explode=myexplode)
plt.show()
```

```python
only_sixes=ball_by_ball_match[ball_by_ball_match.batsman_run.isin([6])]
tot_six_in_match=only_sixes.groupby('ID').agg({'batsman_run':'count'}).sort_values(by='batsman_run',ascending=False)
tot_six_in_match.rename(columns={'batsman_run':'Total_sixes'},inplace=True)
tot_six_in_match.reset_index(inplace=True)
tot_six_in_match.head(5)
```

```python
new_df=pd.merge(tot_six_in_match,Ipl_matches,on='ID')
new_df.head(5)
```
